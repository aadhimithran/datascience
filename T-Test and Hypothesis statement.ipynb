```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male= dataset[dataset['gender']=='M']['salary']
female= dataset[dataset['gender']=='F']['salary']
ttest_ind(male,female)
```

```python
from scipy.stats import ttest_ind
centrals=dataset[dataset['ssc_b']=="Central"]["salary"]
others=dataset[dataset['ssc_b']=="Others"]["salary"]
ttest_ind(centrals,others)
```

#dependent Ttest

```python

from scipy.stats import ttest_rel
hsc=dataset[dataset['hsc_s'] == 'Commerce']['mba_p']
hsc1=dataset[dataset['hsc_s'] == 'Commerce']['etest_p']
# hsc2=dataset[dataset['hsc_s'] == 'Arts']['mba_p'] # t test is for two sample test, more than that need to use anova
ttest_rel(hsc,hsc1)
```

```python
from scipy.stats import ttest_ind
male=dataset[dataset['gender']=='M']["mba_p"]
male1=dataset[dataset['gender']=='M']["etest_p"]
ttest_ind(male,male1)
```

#HYPOTHESIS TESTING

```python
from scipy.stats import ttest_rel
male=dataset[dataset['gender']=='M']["ssc_p"]
male1=dataset[dataset['gender']=='M']["hsc_p"]
ttest_rel(male,male1)
```
